# SVM - Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
X=df.drop('Virus Present', axis=1)
y=df['Virus Present']
```

```python
svc_model=SVC()
```

```python
svc_model.fit(X,y)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_estimator(svc_model,X,y)
```
